```python
import matplotlib.pyplot as plt
import numpy as np

# Target directory
path = 'build'

# Reading the grid parameters
f_grid = open(path + '/grid_params.txt')
params = f_grid.readline().split()
lx = float(params[0])
ly = float(params[1])
nx = int(params[2])
ny = int(params[3])

f_grid.close()

# Reading the solution
z = []
Z = np.empty([100, 100])

f = open(path + '/u.txt')
while True:
    s = f.readline()
    if not s: break
    z.append(float(s.split()[0]))
f.close()

# Plotting
x = np.linspace(0, lx, ny)
y = np.linspace(0, ly, ny)
X, Y = np.meshgrid(x, y)

for i in range(100):
    for j in range(100):
        Z[i, j] = z[i + 100*j]

fig = plt.figure(figsize =(14, 14))
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

# Exporting a .png
plt.savefig('Result.png', dpi=300)
```
